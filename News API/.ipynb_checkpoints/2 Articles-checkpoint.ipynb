## Sentiment Analysis

Analyze the sentiment of the articles (positive, negative, neutral) using Natural Language Processing (NLP) libraries like TextBlob or VADER

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Function to analyze sentiment using TextBlob
def analyze_sentiment_textblob(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Function to analyze sentiment using VADER
def analyze_sentiment_vader(text):
    analyzer = SentimentIntensityAnalyzer()
    vs = analyzer.polarity_scores(text)
    if vs['compound'] > 0.05:
        return 'positive'
    elif vs['compound'] < -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to each article
df['sentiment_textblob'] = df['content'].apply(analyze_sentiment_textblob)
df['sentiment_vader'] = df['content'].apply(analyze_sentiment_vader)

# Display the DataFrame with sentiment analysis results
print(df[['title', 'sentiment_textblob', 'sentiment_vader']])
```

```python
df[['title', 'sentiment_textblob', 'sentiment_vader']]
```

TextBlob
- Positive Sentiment: Assigned when the text contains more words with positive connotations (e.g., "happy," "excellent").
- Negative Sentiment: Assigned when the text contains more words with negative connotations (e.g., "sad," "terrible").

VADER
- Positive Sentiment: Assigned when the composite score is greater than 0.05, indicating the text has a strong positive valence (e.g., "great!").
- Negative Sentiment: Assigned when the composite score is less than -0.05, indicating the text has a strong negative valence (e.g., "horrible").

## Content Analysis
- Analyze the length of articles, the structure, and the most common words or phrases used.
- Identify the most common authors writing on the topic.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from wordcloud import WordCloud
```

```python
# Verify the DataFrame structure
print("DataFrame structure:")
print(df)
print("\nColumns in DataFrame:", df.columns)

# Convert 'publishedAt' column to datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Verify the date conversion
print("\nDataFrame with converted date column:")
print(df)

# Download NLTK stopwords
nltk.download('punkt')
nltk.download('stopwords')

# Text preprocessing function
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
    return filtered_words

# Apply preprocessing to each article
df['processed_content'] = df['content'].apply(preprocess_text)

# Flatten the list of processed words and count the frequency of each word
all_words = [word for content in df['processed_content'] for word in content]
word_freq = Counter(all_words)

# Display the most common keywords
print("Most common keywords:")
print(word_freq.most_common(10))

# Plot the word frequency
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Source analysis
source_counts = df['source'].value_counts()
print("\nMost common sources:")
print(source_counts)

# Author analysis
author_counts = df['author'].value_counts()
print("\nMost common authors:")
print(author_counts)

# Article length analysis
df['content_length'] = df['content'].apply(len)
print("\nArticle length analysis:")
print(df[['title', 'content_length']])

# Trend analysis: Group by day, week, or month
df['month'] = df['publishedAt'].dt.to_period('M')
df['week'] = df['publishedAt'].dt.to_period('W')
df['day'] = df['publishedAt'].dt.to_period('D')

# Group by month
monthly_trend = df.groupby('month').size()
weekly_trend = df.groupby('week').size()
daily_trend = df.groupby('day').size()

# Plotting the trends
plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
monthly_trend.plot(kind='bar', color='skyblue')
plt.title('Monthly Article Trend')
plt.xlabel('Month')
plt.ylabel('Number of Articles')

plt.subplot(3, 1, 2)
weekly_trend.plot(kind='bar', color='lightgreen')
plt.title('Weekly Article Trend')
plt.xlabel('Week')
plt.ylabel('Number of Articles')

plt.subplot(3, 1, 3)
daily_trend.plot(kind='bar', color='salmon')
plt.title('Daily Article Trend')
plt.xlabel('Day')
plt.ylabel('Number of Articles')

plt.tight_layout()
plt.show()
```
