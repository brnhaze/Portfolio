## Remove leading [' and '] from columns

```python
columns_to_rm_brack = ['Based on', 'Box office', 'Budget', 'Cinematography', 'Country', 'Directed by', 'Distributed by', 'Edited by', 'Language', 'Music by', 'Narrated by', 'Produced by', 'Productioncompany', 'Release date', 'Running time', 'Screenplay by', 'Starring', 'Story by', 'Unnamed: 0', 'Written by', 'movie_name']

# function
def clean_column(value):
    if isinstance(value, str) and value.startswith("['") and value.endswith("']"):
        return value[2:-2]
    return value

# apply fuctnion to the specified columns
for column in columns_to_rm_brack:
    df_1[column] = df_1[column].apply(clean_column)

#display cleaned df
df_1.head(1)
```

```python
# Function to convert box office values
def convert_box_office(value):
    # Check for "million" and replace with appropriate number of zeros
    if 'million' in value:
        number = float(re.findall(r'\d+\.?\d*', value)[0])
        return f'{int(number * 1_000_000):,}'
    # Check for "billion" and replace with appropriate number of zeros
    elif 'billion' in value:
        number = float(re.findall(r'\d+\.?\d*', value)[0])
        return f'{int(number * 1_000_000_000):,}'
    return value

# Apply the function to the 'Box office' column
df_2_dated['Box office'] = df_2_dated['Box office'].apply(lambda x: convert_box_office(x) if isinstance(x, str) else x)

# Display the updated DataFrame
df_2_dated
```

```python
# Function to remove parentheses, square brackets, and brackets with spaces
def clean_box_office(value):
    if isinstance(value, str):
        # Remove parentheses, square brackets, and brackets with spaces
        value = re.sub(r'[\(\)\[\]]', '', value)
        value = re.sub(r'\[\s*\]', '', value)
        return value.strip()
    return value

# Apply the function to the 'Box office' column
df_2_dated['Box office'] = df_2_dated['Box office'].apply(clean_box_office)

# Display the updated DataFrame
df_2_dated
```

```python
# Function to add "000000" after numbers if "M" is present
def expand_millions(value):
    if isinstance(value, str) and 'M' in value:
        value = re.sub(r'(\d+\.?\d*)M', lambda m: str(float(m.group(1)) * 1_000_000), value)
    return value

# Apply the function to the 'Box office' column
df_2_dated['Box office'] = df_2_dated['Box office'].apply(expand_millions)

# Display the updated DataFrame
df_2_dated
```

```python
# Function to convert Box office values to float
def convert_to_float(value):
    try:
        # Remove non-numeric characters and convert to float
        return float(re.sub(r'[^\d.]', '', str(value)))
    except ValueError:
        # Return NaN if conversion fails
        return np.nan

# Apply the function to the 'Box office' column
df_2_dated['Box office'] = df_2_dated['Box office'].apply(convert_to_float)
```
